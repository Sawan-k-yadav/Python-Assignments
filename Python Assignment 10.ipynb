1. How do you distinguish between shutil.copy() and shutil.copytree()?

Ans: shutil.copy() copies a single file from one location to another. shutil.copytree() copies an entire directory tree, including all of its subdirectories and files.

Example:

import shutil

### Copy a single file
shutil.copy("myfile.txt", "/new/location")

### Copy a directory tree
shutil.copytree("/old/location", "/new/location")

3. What is the difference between the delete functions in the send2trash and shutil modules?

Ans: The send2trash.move_to_trash() function moves files and directories to the trash bin. The shutil.rmtree() function deletes files and directories permanently.

5. Create a programme that searches a folder tree for files with a certain file extension (such as .pdf
or .jpg). Copy these files from whatever location they are in to a new folder.

Ans: import os
import shutil

def copy_files(source_dir, destination_dir, file_extension):
    """Copies all files with the specified file extension from the source directory to the destination directory.

    Args:
        source_dir: The source directory.
        destination_dir: The destination directory.
        file_extension: The file extension to search for.
    """

    for root, dirs, files in os.walk(source_dir):
        for file in files:
            if file.endswith(file_extension):
                source_path = os.path.join(root, file)
                destination_path = os.path.join(destination_dir, file)
                shutil.copy(source_path, destination_path)

#### Get the source and destination directory paths from the user.
source_dir = input("Enter the source directory path: ")
destination_dir = input("Enter the destination directory path: ")

#### Get the file extension from the user.
file_extension = input("Enter the file extension to search for: ")

#### Copy the files.
copy_files(source_dir, destination_dir, file_extension)

#### Print a success message.
print("All files with the specified file extension have been copied to the destination directory.")
