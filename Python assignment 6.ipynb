To use an escape character, simply type the backslash followed by the character you want to escape. For example, to represent a newline character, you would use the escape character \n.

9. What are the methods for right-justifying, left-justifying, and centering a string?

Ans:The methods for right-justifying, left-justifying, and centering a string are as follows:
Right-justifying: str.rjust(width, fillchar)
Left-justifying: str.ljust(width, fillchar)
Centering: str.center(width, fillchar)
The width argument specifies the width of the string after it has been justified. The fillchar argument specifies the character that will be used to pad the string on the left or right.
