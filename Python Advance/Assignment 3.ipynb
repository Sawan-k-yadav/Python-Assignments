1. What is the concept of an abstract superclass?

Ans: An abstract superclass is a class that is designed to be inherited from, but not instantiated directly. It contains one or more abstract methods, which are methods that have a declaration but no implementation. Subclasses of the abstract superclass must provide implementations for all of the abstract methods.

3. Why does a class need to manually call a superclass&#39;s __init__ method?

Ans:A class needs to manually call a superclass's __init__() method to ensure that the superclass's constructor is called. The superclass's constructor may be used to initialize important state in the superclass.

class Shape:
    def __init__(self, color):
        self.color = color

class Rectangle(Shape):
    def __init__(self, color, width, height):
        super().__init__(color)
        self.width = width
        self.height = height

5. How is the local scope of a class different from that of a function?

ANs:The local scope of a class is the namespace that contains the class's attributes and methods. The local scope of a function is the namespace that contains the function's local variables.

The local scope of a class is different from the local scope of a function in the following ways:

Class attributes and methods are accessible to all instances of the class. Function local variables are only accessible to the function that they are defined in.
Class attributes and methods can be inherited by subclasses of the class. Function local variables cannot be inherited.
Class attributes and methods can be defined outside of the __init__() method. Function local variables must be defined inside the function.

class Counter:
    count = 0

    def increment(self):
        self.count += 1

counter = Counter()
counter.increment()
