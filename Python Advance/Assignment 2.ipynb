Q1. What is the relationship between classes and modules?

Ans:

Classes and modules are two different concepts in Python. A class is a blueprint for creating objects, while a module is a collection of related functions and classes.

Classes can be defined inside modules, but they can also be defined in their own files. Modules can import other modules, but they cannot import classes directly.

Q3. Where and how should be class attributes created?

Ans:

Class attributes are created inside the class definition. They are shared by all instances of the class.

To create a class attribute, 

class MyClass:
    attribute = value

Q4. Where and how are instance attributes created?

Ans:

Instance attributes are created on individual instances of a class. They are not shared by other instances of the class.

To create an instance attribute,

instance = MyClass()
instance.attribute = value

Q5. What does the term &quot;self&quot; in a Python class mean?

Ans:

The self keyword refers to the current instance of the class. It is used to access and modify the instance's attributes and methods.

All methods in a Python class must have a self argument.

Q6. How does a Python class handle operator overloading?

Ans:

Operator overloading is a feature of Python that allows you to define custom behavior for operators when they are used with objects of your own classes.

To overload an operator, you define a special method in your class with the same name as the operator. For example, to overload the addition operator (+), you would define a __add__() method in your class.

Q7. When do you consider allowing operator overloading of your classes?

Ans:

You should consider allowing operator overloading of your classes when it makes sense for your users to be able to use operators with your objects in a natural way.

For example, if you have a class that represents complex numbers, you might want to overload the addition operator so that users can add complex numbers together using the plus sign (+).

Q8. What is the most popular form of operator overloading?

Ans:
The most popular form of operator overloading is the overloading of the comparison operators (<, >, ==, !=, <=, and >=). This allows users to compare objects of your own classes to other objects, including built-in Python types.

For example, if you have a class that represents students, you might want to overload the comparison operators so that users can compare students based on their grades.
