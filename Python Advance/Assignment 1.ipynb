## Q1. What is the purpose of Python's OOP?

The purpose of Python's OOP is to make code more reusable, maintainable, and extensible. OOP allows us to model real-world objects and their relationships in code, which makes it easier to understand and write complex programs.

## Q3. How do we distinguish between a class object and an instance object?

A class object is a blueprint for creating instances of the class. It contains the class's attributes and methods. An instance object is a concrete representation of the class. It has its own unique set of values for the class's attributes.

To distinguish between a class object and an instance object, we can use the type() function. The type() function returns the type of an object. For a class object, the type() function will return the class itself. For an instance object, the type() function will return the class that created the instance object.

## Q4. What makes the first argument in a class's method function special?

The first argument in a class's method function is a special argument called self. The self argument represents the instance of the class on which the method is being called.

All methods in a class must have a self argument. This allows methods to access and modify the instance's attributes.

## Q5. What is the purpose of the init method?

The __init__ method is a special method that is called when a new instance of a class is created. It is used to initialize the instance's attributes.

The __init__ method must have at least one argument, which is the self argument.

## Q6. What is the process for creating a class instance?

To create a class instance, we use the __new__() and __init__() methods.

The __new__() method is used to create a new instance of the class. It returns a reference to the new instance.

The __init__() method is used to initialize the instance's attributes. It is called after the instance has been created by the __new__() method.

The following code shows how to create a class instance:

Python
class Person:
    def __init__(self, name):
        self.name = name

#### Create a new instance of the Person class
person = Person("Sawan")

#### Print the person's name
print(person.name)
Output:

Sawan

## Q7. What is the process for creating a class?

To create a class, we use the class keyword. The following code shows how to create a simple class:

Python
class Person:
    pass

This code creates a class called Person. The pass statement is used to indicate that the class does not have any attributes or methods yet.

we can add attributes and methods to a class by defining them inside the class definition. For example, the following code adds a name attribute and a greet() method to the Person class:

Python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, my name is {self.name}")

#### Create a new instance of the Person class
person = Person("Sawan")

#### Call the greet() method on the person instance
person.greet()

Output:

Hello, my name is Sawan
