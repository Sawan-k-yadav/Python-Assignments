6. What are the five levels of logging?

Ans:
The five levels of logging are:

DEBUG: The most detailed level of logging.
INFO: Logs general information about the program's execution.
WARNING: Logs potential problems with the program's execution.
ERROR: Logs errors that occurred during the program's execution.
CRITICAL: Logs critical errors that prevent the program from continuing

8.Why is using logging messages better than using print() to display the same message?

Ans: Logging messages are better than print() for a few reasons:

Logging messages can be sent to different destinations, such as files, the console, or email. Print() messages can only be sent to the console.
Logging messages can be configured to different levels of detail. This allows you to control how much information is logged. Print() messages are always printed to the console, regardless of the level of detail.
Logging messages can be easily filtered and searched. Print() messages cannot be filtered or searched.


9. What are the differences between the Step Over, Step In, and Step Out buttons in the debugger?

Ans:
Step Over: Steps over the current line of code. If the current line of code is a function call, the debugger will not step into the function.
Step In: Steps into the current line of code. If the current line of code is a function call, the debugger will step into the function.
Step Out: Steps out of the current function call.

11. What is the concept of a breakpoint?

Ans:

A breakpoint is a marker that you can set in your code to tell the debugger to stop at that location. Once you have set a breakpoint, you can click the Continue button to run your program until it reaches the breakpoint. At that point, the debugger will stop and allow you to examine the state of your program.
