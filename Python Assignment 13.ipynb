2.What do you pass to csv.reader() and csv.writer() to create reader and writer objects?

Ans: 
To create a reader object, we pass a file object to the csv.reader() function. To create a writer object, you pass a file object and a dialect to the csv.writer() function. A dialect is a set of rules for how to read and write CSV data.

For example, the following code creates a reader object for a CSV file called example.csv:


import csv

with open("example.csv", "r") as f:
    reader = csv.reader(f)


The following code creates a writer object for a CSV file called example.csv with the excel dialect:


import csv

with open("example.csv", "w") as f:
    writer = csv.writer(f, dialect="excel")
    
*if we will not use dialect then it will use default dialect

3. What modes do File objects for reader and writer objects need to be opened in?

Ans:
The file object for a reader object needs to be opened in read mode ("r"). The file object for a writer object needs to be opened in write mode ("w") or append mode ("a").

4. What method takes a list argument and writes it to a CSV file?

Ans:
The writer.writerow() method takes a list argument and writes it to a CSV file. The list should contain the values for each row in the CSV file.

For example, the following code writes the list ["name", "age"] to the CSV file example.csv:


import csv

with open("example.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(["name", "age"])

5. What do the keyword arguments delimiter and line terminator do?

Ans:
The delimiter keyword argument specifies the character that is used to separate the values in each row of the CSV file. The default delimiter is a comma.

The line terminator keyword argument specifies the character or sequence of characters that is used to terminate each row of the CSV file. The default line terminator is a newline character ("\n").
