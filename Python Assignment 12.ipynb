1. In what modes should the PdfFileReader() and PdfFileWriter() File objects will be opened?

Ans: PdfFileReader() file objects should be opened in rb -> read binary mode,

and PdfFileWriter() file objects should be opened in wb -> write binary mode

2. From a PdfFileReader object, how do you get a Page object for page 5?

Ans:
pdf_reader = PyPDF2.PdfFileReader(open("example.pdf", "rb"))

page_5 = pdf_reader.getPage(4)

The getPage() method takes an integer argument, which is the index of the page that we want to get. The first page of the PDF document has index 0.

3. What PdfFileReader variable stores the number of pages in the PDF document?

Ans: The PdfFileReader variable that stores the number of pages in the PDF document is numPages. This variable is an integer.

4. If a PdfFileReader object’s PDF is encrypted with the password swordfish, what must you do
before you can obtain Page objects from it?

Ans: We must call the decrypt() method on the PdfFileReader object before we can obtain Page objects from it.
Example:

pdf_reader = PyPDF2.PdfFileReader(open("example.pdf", "rb"))

pdf_reader.decrypt("swordfish")

page_5 = pdf_reader.getPage(4)

5. What methods do you use to rotate a page?

Ans: To rotate a page, you can use the rotateClockwise() or rotateCounterClockwise() methods on the Page object. 
For example:

page_5.rotateClockwise(90)

6. What is the difference between a Run object and a Paragraph object?

Ans:The difference between a Run object and a Paragraph object is that a Paragraph object represents a paragraph of text, while a Run object represents a contiguous group of characters within a paragraph. Run objects can be used to format text, such as making it bold, italic, or underlined. Paragraph objects can be used to format paragraphs, such as setting the alignment and margins.
New Paragraph objects can be added only to the end of the document, and new Run objects can be added only to the end of a Paragraph object.

8. What type of object has bold, underline, italic, strike, and outline variables?

Ans: The type of object that has bold, underline, italic, strike, and outline variables is a Run object. These variables can be used to format the text of the Run object.

9. What is the difference between False, True, and None for the bold variable?

Ans: False: The Run object will not be bold.

True: The Run object will always be bold, regardless of the style's bold setting.

None: The Run object will use the style's bold setting.

variable named doc?

doc.save('hellothere.docx')

12. What integers represent the levels of headings available in Word documents?

Ans:
The levels of headings available in Word documents are represented by the following integers:

0: Normal

1: Heading 1

2: Heading 2

3: Heading 3

4: Heading 4

5: Heading 5

6: Heading 6

7: Heading 7

8: Heading 8

9: Heading 9

You can use these integers to set the heading level of a paragraph using the Paragraph.style property. 

For example, the following code shows how to set the heading level of a paragraph to Heading 1:


paragraph = doc.add_paragraph("This is a heading")

paragraph.style = "Heading 1"
